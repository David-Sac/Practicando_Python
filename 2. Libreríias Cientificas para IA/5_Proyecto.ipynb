# Proyecto: Análisis de Ventas

Datos de ventas de una tienda (producto, precio unitario, cantidad y mes)
El objetivo es:
- Calcular totales y estadísticas.
- Filtrar y analizar productos más vendidos.
- Visualizar los resultados con gráficos.

```python
import pandas as pd

datos = {
    "Producto": ["Arroz", "Azúcar", "Papa", "Cebolla", "Leche", "Pan", "Arroz", "Papa", "Leche", "Pan"],
    "Precio": [5.0, 4.5, 2.0, 3.0, 4.0, 1.5, 5.0, 2.0, 4.0, 1.5],
    "Cantidad": [20, 35, 50, 25, 40, 60, 30, 70, 80, 90],
    "Mes": ["Enero", "Enero", "Enero", "Febrero", "Febrero", "Febrero", "Marzo", "Marzo", "Abril", "Abril"]
}

df = pd.DataFrame(datos)
print(df.head())
```

## Análisis de ventas por fila

Calcular total de ventas por fila

```python
df["Total"] = df["Precio"] * df["Cantidad"]
print(df)
```

Estadísticas generales

```python
print("Promedio de precios:", df["Precio"].mean())
print("Total de ventas:", df["Total"].sum())
print("Producto mas vendido:")
print(df.groupby("Producto")["Total"].sum().sort_values(ascending=False).head(3))
```

Venta por mes

```python
print("Ventas por mes:")
print(df.groupby("Mes")["Total"].sum())
```

## Visualizacion con Matplotlib

Gráfico de barras: ventas por producto

```python
import matplotlib.pyplot as plt
df.groupby("Producto")["Total"].sum().plot(kind="bar", color="skyblue")
plt.title("Ventas por mes")
plt.ylabel("Soles")
plt.show()
```

Gráfico de línea: ventas por mes

```python
df.groupby("Mes")["Total"].sum().plot(kind="line", marker="o", color="green")
plt.title("Ventas por mes")
plt.ylabel("Soles")
plt.show()
```

Gráfico de torta: participación por producto

```python
df.groupby("Producto")["Total"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Participación de productos en las ventas")
plt.ylabel("")
plt.show()
```
