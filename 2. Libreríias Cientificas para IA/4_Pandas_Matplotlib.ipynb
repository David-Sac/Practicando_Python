# Pandas & Matplotlib

Es la librería másusada en Python para graficar datos.
Pandas tiene una intefracion directa: cualquier DataFrame o Series puede graficarse con .plot() y usa Matplotlib por detras

## Gráficos básicos con Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Producto": ["Arroz", "Azucar","Papa","Cebolla"],
    "Precio": [10,20,30,15],
    "Cantidad": [4,14,16,10]
})

#Gráfico de barras
df.plot(kind="bar", x="Producto", y="Cantidad", legend=False)
plt.title("Cantidad de productos")
plt.ylabel("Unidades")
plt.show
```

## Tipos de gráficos comunes

```python
# Linea
df["Precio"].plot(kind="line", marker="o", title="Precio de productos")
plt.show()
```

```python
# Barras
df.plot(kind="bar", x="Producto", y="Precio")
plt.show()
```

```python
# Barras Horizontales
df.plot(kind="barh", x="Producto", y="Cantidad")
plt.show()
```

```python
# Histograma
df["Precio"].plot(kind="hist", bins=5, title="Distribucion de precios")
plt.show()
```

```python
# Gráfico de torta
df.set_index("Producto")["Cantidad"].plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("") # quitar texto de eje
plt.title("Participacion de productos")
plt.show
```

## Personalización con Matplotlib

Con `plt` puedes añadir detalles:

```python
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Mi gráfico")
plt.grid(True)
```

## Ejemplo: Ventas por mes

```python
ventas = pd.DataFrame({
    "Mes": ["Enero","Febrero","Marzo","Abril"],
    "Ventas": [1000,2000,900,1400]
})

ventas.plot(kind="line", x="Mes", y="Ventas", marker="o")
plt.title("Ventas por mes")
plt.ylabel("Monto en soles")
plt.show()
```

## Mini proyecto

Análisis de notas y visualización

```python
df = pd.DataFrame({
    "Alumno": ["David","Ana","Luis","Carla","Juan"],
    "Curso": ["Matematicas","Matematicas","Historia","Historia","Matematicas"],
    "Nota": [18,15,12,14,19]
})

# Promedio por curso
promedios = df.groupby("Curso")["Nota"].mean()

# Gráficos de barra de promedios
promedios.plot(kind="bar", color="skyblue")
plt.title("Promedio de notas por curso")
plt.ylabel("Nota")
plt.show()
```
